# Patch manipulation routines

This notebook contains commonly used routines for patch-based image manipulation: 
- On the fly patch extraction
- Data augmentation (rotation, flipping)
- Patch aggregation/merging

## Import the libraries

The Scikit-Image library is used for all the image processing routines.

```python
import os
import tensorflow as tf
import numpy as np
import pandas as pd
import functools
rng = np.random.seed(100)
 
from skimage import io, transform, exposure, filters, color
from plotting import plot_images, plt
%pylab inline
```

## Read and process paths

```python
def read_df(fpath, data_dir=''):
    """
    Build an array of image paths along with their corresponding segmentation paths.
    
    Params:
    - fpath: string,
        path to a file containing two columns: an `im_name` column and `gt_name` one.
        example of files are provided in datasets/DRIVE
    - data_dir: string, optional
        the dataset directory to prepend to the paths in `fpath` in case they are not absolute paths.
    
    Returns:
    - x_paths: array
        the list of image paths of the dataset
    - y_paths: array,
        the corresponding segmentation paths
    
    """
    df_train = pd.read_csv(fpath)
    x_paths = df_train['im_name'].map(lambda s: os.path.join(data_dir,s))
    y_paths = df_train['gt_name'].map(lambda s: os.path.join(data_dir,s))
    return x_paths, y_paths

def _process_pathnames(fname, lname, resize=None): 
    """
    Read and resize an image and its associated segmentation.
    
    Params:
    - fname: string,
        path to an image
    - lname: string,
        path to a segmentation
    - resize: tuple (Height, Width), optional
        if not None, the image and annotation are resized.
    
    Returns:
    - img: the image
    - gt: the segmentation
    """
    img = io.imread(fname)
    gt = io.imread(lname)
    if gt.ndim < 3:
        gt  = np.expand_dims(gt, -1)
    gt = gt[...,:1]
    gt = (gt > 0).astype(int) # binarize the ground-truth
    if resize is not None:
        img = transform.resize(img, resize)
        gt = transform.resize(gt, resize)
        gt = gt >= filters.threshold_otsu(gt)
    return img, gt 
```

## Data augmentation routines

```python
def shift_img(img, gt, width_shift_range, height_shift_range, rotate_range): 
    """
    Applies random translation and rotation to a given image and annotation
    
    Params:
    - img: input image
    - gt: associated annotation
    - width_shift_range: float,
        random translation range to apply horizontally
    - height_shift_range: float,
        random translation range to apply vertically
    - rotate_range: float,
        random rotation range to apply
    
    Returns:
    - img: shifted image
    - gt: shifted annotation
    
    """
    if width_shift_range  or height_shift_range:
        if width_shift_range:
            width_shift_range = np.random.uniform(-width_shift_range * img.shape[1],
                                                   width_shift_range * img.shape[1])
        if height_shift_range:
            height_shift_range = np.random.uniform(-height_shift_range * img.shape[0],
                                                   height_shift_range * img.shape[0]) 
        tr = transform.AffineTransform(translation=(width_shift_range, height_shift_range )) 
        img = transform.warp(img, tr, preserve_range=True)
        gt   = transform.warp(gt, tr, preserve_range=True)
        
    if rotate_range :
        if isinstance(rotate_range, np.ScalarType):
            degre = np.random.uniform(-rotate_range,rotate_range)
        else:
            degre = np.random.uniform(rotate_range[0], rotate_range[1])
        img = transform.rotate(img, degre, preserve_range=True)
        gt  = transform.rotate(gt, degre, preserve_range=True)
        
    return img, gt

def flip_img(img, gt, horizontal_flip, vertical_flip):
    """
    Applies randomn flips to a given image and associated annotation.
    """
    if horizontal_flip:
        flip_prob = np.random.uniform(0.0, 1.0)
        img, gt = (img, gt) if flip_prob >= 0.5 else (np.flip(img, 1), np.flip(gt, 1))
    if vertical_flip:
        flip_prob = np.random.uniform(0.0, 1.0)
        img, gt = (img, gt) if flip_prob >= 0.5 else (np.flip(img, 0), np.flip(gt, 0))
    return img, gt

def _process_imgt(img, gt, gamma=0, gray=False, xyz=False, hed=False, clahe=False,
            horizontal_flip=False, vertical_flip=0, width_shift_range=0,
            height_shift_range=0, rotate_range=(0,0)):
    """
    Applies some color normalization on the input image and
    randomnly shifts it along with its annotation.
    
    Params:
    - img: input image
    - gt: associated annotation
    - gamma: float, default 0,
        gamma correction value
    - gray: bool, default false,
        whether or not to convert the image to gray-scale
    - xyz: bool, default false,
        whether or not to convert the image to xyz color space
    - hed: bool, default false,
        whether or not to convert the image to hed color space
    - clahe: bool, default false,
        whether or not to apply CLAHE normalization
    - horizontal_flip: bool, default false,
        whether or not to randomnly flip the images horizontally
    - vertical_flip: bool, default false,
        whether or not to randomnly flip the images vertically
    - height_shift_range: float, default 0,
        random translation range to apply vertically
    - width_shift_range: float, default 0,
        random translation range to apply horizontally
    - rotate_range: float,
        random rotation range to apply
    
    Returns:
    - img: shifted image
    - gt: shifted annotation
    """
    img = exposure.rescale_intensity(img.astype(float), out_range=(0,1))
    if gray:
        img = color.rgb2gray(img)
    if xyz:
        img = color.rgb2xyz(img)
    if hed:
        img = color.rgb2hed(img)
    img = exposure.rescale_intensity(img, out_range=(0,1))
    if clahe:
        img = exposure.equalize_adapthist(img)
    if gamma:
        img = exposure.adjust_gamma(img, gamma)
        img = exposure.rescale_intensity(img, out_range=(0,1))
    if img.ndim == 2:
        img = np.expand_dims(img, -1)

    img, gt = flip_img(img, gt, horizontal_flip, vertical_flip)    
    img, gt = shift_img(img, gt, width_shift_range, height_shift_range, rotate_range) 
    img = exposure.rescale_intensity(img, out_range=(0,1)) 

    return img, gt
```

## Patch position extration

```python
def fixed_patch_ids_creation(im_paths, gt_paths, spatial_shape=None,
                            p_stride=16, per_label=0, shuffle=True, mask=None):
    """
    Creates a list of the form: 
    ['patch position in th image (x_index, y_index)', 'image path', 'annotation path']. 
    
    Params:
    - im_paths: list of image paths
    - gt_paths: list of annotation paths
    - spatial_shape:  tuple (Height, Width), optitional
        To resize the annotations before patch position extraction in case
        the extraction is performed based on the class distribution.
    - p_stride: int, default 16
        if non null, strided patch position are extracted for each image path
    - per_label: int, default 0,
        if non null, a `per_label` number of patches are extracted for each label in the annotations.
    - shuffle: bool, default True,
        whether or not to shuffle the extracted positions.
    - mask: array, optional, must have the same shape as the annotations
        if given, the patch positions are extracted where `mask` is not null.
    
    Returns:
    - patch_position_dataset: list of the form ['patch position', 'image path', 'annotation path']. 
    """
    all_ids = []
    mask = mask if mask is not None else 1
    for im_path, gt_path in zip(im_paths, gt_paths) : 
        if p_stride > 0:
            ids = np.zeros(spatial_shape, dtype='int')
            if ids.ndim == 2:
                ids[0::p_stride, 0::p_stride] = 1   
            else: 
                ids[0::p_stride, 0::p_stride, 0::p_stride] = 1   
            ids = ids * mask
            ids = np.array(np.nonzero(ids)).T
            n  = len(ids)
            ap = np.c_[[im_path]*n, [gt_path]*n, ids] 
            all_ids.extend(ap) 
        
        if per_label >0: 
            # Adding samples based on the classes distribution.
            _, gt = _process_pathnames(im_path, gt_path, resize=spatial_shape) 
            cls_ids = []
            for c in np.unique(gt):
                search_area = np.nonzero((np.squeeze(gt) == c) * mask)
                if len(search_area[0]) == 0:
                    continue
                search_area = np.array(search_area).T
                search_area = np.random.permutation(search_area) 
                cls_ids.append(search_area[:per_label])
            cls_ids = np.concatenate([x for x in cls_ids])  
            n  = len(cls_ids)
            ap = np.c_[[im_path]*n, [gt_path]*n, cls_ids] 
            all_ids.extend(ap) 

    all_ids = np.array(all_ids)
    if shuffle:
        np.random.shuffle(all_ids)  
    return all_ids 

```

## Building the patch generator (using the Keras.Sequence class)

```python

class Patch_Sequence(tf.keras.utils.Sequence):
    """
    On-the-fly patch extractor.
    
    Params:
    - fixed_patch_ids: list obtained from `fixed_patch_ids_creation`
    - p_shape: tuple,
        patch shape
    - reader_fn: func,
        takes as input two paths and produced to arrays (image and segmentation)
    - prepro_fn: func,
        takes as inputs an image and segmentation and preprocess them (data augmentation and color normalization)
    - batch_size: int,
        the batch size
    - MAX_IM_QUEUE: int, default 50
        To avoid multiple image reading preprocessing and padding, images are stored in 
        a stack. The variable defines the maximum number of images to store in the stack.
    """
    def __init__(self, fixed_patch_ids, p_shape=(32,32,3),
                reader_fn=functools.partial(_process_pathnames),
                preproc_fn=functools.partial(_process_imgt),
                batch_size=32,
                MAX_IM_QUEUE=50):
        self.ids = fixed_patch_ids #
        self.p_shape = p_shape
        self.batch_size = batch_size
        self.reader_fn = reader_fn
        self.preproc_fn = preproc_fn
        self.MAX_IM_QUEUE = MAX_IM_QUEUE
        self.im_stack = {}

    def __len__(self):
        return int(np.ceil(len(self.ids) / float(self.batch_size)))

    def __getitem__(self, idx):
        cur_id = self.ids[idx * self.batch_size:(idx + 1) * self.batch_size] 

        batch_x = []
        batch_y = []
        batch_place = []
        for pos in cur_id: 
            pid, pim, pgt = pos[2:], pos[0], pos[1] 
            x_p, y_p = pid.astype(int) 
            hash_im = hash(pim)
            if not self.im_stack.has_key(hash_im):
                img, gt = self.reader_fn(pim, pgt) 
                img, gt = self.preproc_fn(img, gt) 
                img = np.pad(img, ((self.p_shape[0]//2,), (self.p_shape[1]//2,), (0,)), mode='reflect')
                gt = np.pad(gt, ((self.p_shape[0]//2,), (self.p_shape[1]//2,), (0,)), mode='reflect')  
                if len(self.im_stack.keys()) > self.MAX_IM_QUEUE:
                    self.im_stack.popitem()
                self.im_stack[hash_im] = (img, gt)
            else:
                img, gt = self.im_stack[hash_im] 
            patch = img[x_p:x_p+self.p_shape[0], y_p:y_p+self.p_shape[1]] 
            label = gt[x_p:x_p+self.p_shape[0], y_p:y_p+self.p_shape[1]] 
            place_im = img.copy()
            place_im[x_p:x_p+self.p_shape[0], y_p:y_p+self.p_shape[1]] = 10
            place_label = gt.copy()
            place_label[x_p:x_p+self.p_shape[0], y_p:y_p+self.p_shape[1]] = 10
            batch_x.append(patch)
            batch_y.append(label)
            batch_place.append([place_im, place_label])
            
        return np.array(batch_x), np.array(batch_y), batch_place
    
    def on_epoch_end(self, epoch=None, logs=None):
        self.im_stack = {}
        
def get_gen(dataset_ids, p_shape, batch_size=1, gamma=0., 
            clahe=False,  gray=False, xyz=False, hed=False,
            width_shift_range=0, height_shift_range=0, 
            horizontal_flip=False,vertical_flip=False,
            rotate_range=0, resize=None,
            MIN_PATCH_STD=None, MAX_IM_QUEUE=100,
            return_keras_queuer=False):
    """
    Partially build the reader function and the preprocessing one to be used in the patch extraction class.
    """
    prepro_cfg = dict(gamma=gamma, horizontal_flip=horizontal_flip,
                    vertical_flip=vertical_flip, width_shift_range=width_shift_range,
                    height_shift_range=height_shift_range, clahe=clahe, gray=gray, xyz=xyz, hed=hed)
    prepro_fn = functools.partial(_process_imgt, **prepro_cfg) 
    reader_cfg = dict(resize=resize)
    reader_fn = functools.partial(_process_pathnames, **reader_cfg)
 
    return Patch_Sequence(dataset_ids, p_shape=p_shape,
                    reader_fn=reader_fn, preproc_fn=prepro_fn,
                    batch_size=batch_size, MAX_IM_QUEUE=MAX_IM_QUEUE)
```

## Patch merging

- Extract patches from the first image path: X[0]
- Build the patch generator

```python
patches_positions = fixed_patch_ids_creation([X[0]],[Y[0]], spatial_shape=image_shape[:2], p_stride=32, shuffle=False)
patch_gen = get_gen(patches_positions, p_shape, batch_size=1, 
                    gamma=gamma, horizontal_flip=0, width_shift_range=0.0,
                    height_shift_range=0.0, vertical_flip=0, rotate_range=0,
                    clahe=clahe, gray=gray, resize=image_shape[:2],) 
```

- Prepare the output array
- Create an array that counts the number of patches used for each pixel: to average the result afterward.

```python
merge = np.zeros(image_shape[:2]+(1,), dtype=float)
merge = np.pad(merge, ((p_shape[0]//2,), (p_shape[1]//2,), (0,)), mode='reflect')
cnt   = np.zeros_like(merge) 
```

- Simple patch aggregation loop
- Average the prediction
- Handle the padding pixels

```python
for p  in range(len(patch_gen)):
    patch = patch_gen[p][1][0]
    x_p, y_p = patches_positions[p][2:].astype(int)    
    merge[x_p:x_p+p_shape[0], y_p:y_p+p_shape[1]] += patch
    cnt[x_p:x_p+p_shape[0], y_p:y_p+p_shape[1]] += 1 

merge[cnt!=0] = merge[cnt!=0]/cnt[cnt!=0] 
merge = merge[p_shape[0]//2:-p_shape[0]//2, p_shape[0]//2:-p_shape[0]//2]  
```
